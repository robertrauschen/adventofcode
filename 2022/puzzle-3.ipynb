```python
INPUTFILE = 'input-3.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
rows = (input.split("\n"))
```

```python
rucksacks = []
for row in rows:
  n = len(row)
  rucksacks.append((row[0:n//2], row[n//2:]))
```

```python
sum = 0
for rucksack in rucksacks:
  for char in rucksack[0]:
    if char in rucksack[1]:
      common = char
  if common.islower():
    prio = ord(common) - 96
  else:
    prio = ord(common) - 38
  sum += prio
  #print(sum)
print(sum)
```

```python
groups = []
counter = 0
for row in rows:
  if counter == 0:
    group = []
  group.append(row)
  counter += 1
  if counter == 3:
    groups.append(group)
    counter = 0
```

```python
sum = 0
for group in groups:
  for char in group[0]:
    if char in group[1]:
      if char in group[2]:
        common = char
  if common.islower():
    prio = ord(common) - 96
  else:
    prio = ord(common) - 38
  sum += prio
  #break
print(sum)
```
