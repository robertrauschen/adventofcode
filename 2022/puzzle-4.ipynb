```python
INPUTFILE = 'input-4.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
rows = (input.split("\n"))
```

```python
pairs = []
for row in rows:
  pair = row.split(",")
  pairs.append((pair[0].split("-"), pair[1].split("-")))
```

```python
count = 0
for pair in pairs:
  if int(pair[0][0]) >= int(pair[1][0]) and int(pair[0][1]) <= int(pair[1][1]) or int(pair[0][0]) <= int(pair[1][0]) and int(pair[0][1]) >= int(pair[1][1]):
    #print("counts")
    count += 1
  #print(pair, count)
```
