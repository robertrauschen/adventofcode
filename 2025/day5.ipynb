```python
INPUTFILE = 'input5.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
```

```python
ranges, ids = input.split("\n\n")
ranges = ranges.split("\n")
ids = ids.split("\n")
```

```python
num_ranges = []
for str_range in ranges:
    left, right = str_range.split("-")
    num_ranges.append(range(int(left), int(right)+1))
```

```python
score = 0
for id in ids:
    if any(int(id) in num_range for num_range in num_ranges):
        score += 1
print(score)
```

```python
tup_ranges = []
for str_range in ranges:
    left, right = map(int, str_range.split("-"))
    tup_ranges.append((left, right))
```

```python
merge = True
while merge:
    merge = False
    new_tup_ranges = tup_ranges
    for i in range(len(tup_ranges)):
        left, right = tup_ranges[i]
        for j in range(len(tup_ranges)):
            if i == j:
                continue
            left2, right2 = tup_ranges[j]
            if left <= right2 <= right and left2 <= left:
                merge = True
                new_tup_ranges[j] = (left2, right)
                break
            elif left <= left2 <= right and right2 >= right:
                merge = True
                new_tup_ranges[j] = (left, right2)
                break
            elif left2 <= left and right2 >= right:
                merge = True
                break
        if merge:
            new_tup_ranges.pop(i)
            break
    tup_ranges = new_tup_ranges
```

```python
score = 0
for tup in tup_ranges:
    score += tup[1] - tup[0] + 1
print(score)
```
