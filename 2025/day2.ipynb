```python
import re

INPUTFILE = 'input2.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
```

```python
ranges = input.split(",")
```

```python
def count_digits(n: int) -> int:
    count = 1
    if n >= 100000000:
        count += 8
        n //= 100000000
    if n >= 1e4:
        count += 4
        n //= 10000
    if n >= 100:
        count += 2
        n //= 100
    if n >= 10:
        count += 1
    return count
```

```python
score = 0
for ra in ranges:
    r1, r2 = ra.split("-")
    for id in range(int(r1), int(r2)+1):
        digits = count_digits(id)
        if digits % 2 == 0:
            first = id // (10 ** (digits // 2))
            second = id % (10 ** (digits // 2))
            if first == second:
                score += id
print(f"Final score: {score}")
```

```python
score = 0
for ra in ranges:
    r1, r2 = ra.split("-")
    for id in range(int(r1), int(r2)+1):
        str_id = str(id)
        digits = len(str_id)
        if digits % 2 == 0:
            first = str_id[:digits // 2]
            second = str_id[digits // 2:]
            if first == second:
                score += id
print(f"Final score: {score}")
```

```python
score = 0
for ra in ranges:
    r1, r2 = ra.split("-")
    for id in range(int(r1), int(r2)+1):
        digits = count_digits(id)
        for frag in range(2, digits + 1):
            if digits % frag == 0:
                part_size = digits // frag
                parts = []
                temp_id = id
                for _ in range(frag):
                    part = temp_id % (10 ** part_size)
                    parts.append(part)
                    temp_id //= (10 ** part_size)
                if all(p == parts[0] for p in parts):
                    score += id
                    break
print(f"Final score: {score}")
```

```python
score = 0
for ra in ranges:
    r1, r2 = ra.split("-")
    for id in range(int(r1), int(r2)+1):
        digits = count_digits(id)
        for frag in range(2, digits + 1):
            if digits % frag == 0:
                part_size = digits // frag
                prev_part = None
                temp_id = id
                valid = False
                for _ in range(frag):
                    part = temp_id % (10 ** part_size)
                    if prev_part is not None:
                        if part != prev_part:
                            valid = True
                            break
                    prev_part = part
                    temp_id //= (10 ** part_size)
                if not valid:
                    score += id
                    break
print(f"Final score: {score}")
```
