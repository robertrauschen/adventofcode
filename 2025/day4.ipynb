```python
INPUTFILE = 'input4.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
lines = input.split("\n")
```

```python
new_lines = ["" for _ in range(len(lines))]
score = 0
for i in range(len(lines)):
    for j in range(len(lines[i])):
        if lines[i][j] == ".":
            new_lines[i] += "."
            continue
        count = 0
        if i > 0:
            if lines[i-1][j] == "@":
                count += 1
            if j > 0:
                if lines[i-1][j-1] == "@":
                    count += 1
            if j < len(lines[i]) - 1:
                if lines[i-1][j+1] == "@":
                    count += 1
        if i < len(lines) - 1:
            if lines[i+1][j] == "@":
                count += 1
            if j > 0:
                if lines[i+1][j-1] == "@":
                    count += 1
            if j < len(lines[i]) - 1:
                if lines[i+1][j+1] == "@":
                    count += 1
        if j > 0:
            if lines[i][j-1] == "@":
                count += 1
        if j < len(lines[i]) - 1:
            if lines[i][j+1] == "@":
                count += 1
        if count < 4:
            score += 1
            new_lines[i] += "x"
        else:
            new_lines[i] += lines[i][j]

print(score)
```
