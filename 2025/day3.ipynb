```python
import re

INPUTFILE = 'input3.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
lines = input.split("\n")
```

```python
score = 0
for line in lines:
    max1 = 0
    ind1 = []
    for i in range(len(line)):
        digi = int(line[i])
        if digi > max1:
            max1 = digi
            ind1 = [i]
        elif digi == max1:
            ind1.append(i)
    if len(ind1) > 1:
        score += max1 * 11
    else:
        max_i = ind1[0]
        if max_i == len(line) - 1:
            max2 = int(max(line[:-1]))
            score += max2 * 10 + max1
        else:
            max2 = int(max(line[max_i+1:]))
            score += max1 * 10 + max2
    
print(score)
```

Part 2

```python
score = 0
for line in lines:
    rem_dig = 12
    on_bat = ""
    left_lim = 0
    while rem_dig > 0:
        right_lim = len(line) - rem_dig
        curr_max = 0
        curr_max_i = 0
        for i in range(left_lim, right_lim + 1):
            digi = int(line[i])
            if digi > curr_max:
                curr_max = digi
                curr_max_i = i
        left_lim = curr_max_i + 1
        on_bat += line[curr_max_i]
        rem_dig -= 1
    # print(on_bat)
    score += int(on_bat)
    
print(score)
```
