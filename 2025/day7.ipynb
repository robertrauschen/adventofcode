```python
INPUTFILE = 'input7.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
lines = input.split("\n")
```

```python
score = 0
beam = {lines[0].find("S")}
for line in lines[1:]:
    for beam_pos in beam.copy():
        if line[beam_pos] == "^":
            beam.remove(beam_pos)
            beam.add(beam_pos - 1)
            beam.add(beam_pos + 1)
            score += 1
print(score)
```

Part 2

```python
beam = {lines[0].find("S"): 1}
for line in lines[1:]:
    for beam_pos in beam.copy():
        if beam[beam_pos] and line[beam_pos] == "^":            
            if beam_pos - 1 in beam:
                beam[beam_pos - 1] += beam[beam_pos]
            else:
                beam[beam_pos - 1] = beam[beam_pos]
            if beam_pos + 1 in beam:
                beam[beam_pos + 1] += beam[beam_pos]
            else:
                beam[beam_pos + 1] = beam[beam_pos]

            beam[beam_pos] -= beam[beam_pos]
                
print(sum(beam.values()))
```
