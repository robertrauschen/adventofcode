```python
import re

INPUTFILE = 'input1.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
rows = (input.split("\n"))
```

```python
pos = 50
score = 0
for row in rows:
    # print(row)
    if row[0] == 'L':
        fac = -1
    elif row[0] == 'R':
        fac = 1
    else:
        raise ValueError(f"Invalid direction in row {row}")
    step = int(row[1:])
    pos += fac * step
    pos = pos % 100
    # print(f"Position: {pos}")
    if pos == 0:
        score += 1
print(f"Final score: {score}")        
```
