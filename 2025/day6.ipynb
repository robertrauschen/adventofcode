```python
INPUTFILE = 'input6.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
```

```python
data = []
for line in input.split("\n"):
    data.append([item for item in line.split(" ") if item])

score = 0
for i in range(len(data[0])):
    if data[-1][i] == "*":
        res = 1
        for j in range(len(data)-1):
            res *= int(data[j][i])
        score += res
    elif data[-1][i] == "+":
        res = 0 
        for j in range(len(data)-1):
            res += int(data[j][i])
        score += res
    else:
        raise ValueError("Invalid operator")
    
print(score)
```

Part 2

```python
lines = input.split("\n")
data = ["" for _ in range(len(lines[0]))]
for i in range(len(lines[0])):
    for j in range(len(lines)-1):
        data[i] += lines[j][i]
```

```python
ops = [op for op in lines[-1] if op in ["*", "+"]]
```

```python
packs = [[] for _ in range(len(ops))]
index = 0
for line in data:
    if line.strip():
        packs[index].append(int(line))
    else:
        index += 1
```

```python
score = 0
for op, pack in zip(ops, packs):
    if op == "*":
        res = 1
        for item in pack:
            res *= item
        score += res
    if op == "+":
        score += sum(pack)
print(score)
```
