```python
INPUTFILE = 'input8.txt'

with open(INPUTFILE, 'r') as file:
    input = file.read()
lines = input.split("\n")
```

```python
def calc_dist(a: tuple, b:tuple):
    return (
        (a[0] - b[0])**2
        + (a[1] - b[1])**2
        + (a[2] - b[2])**2
    )    
```

```python
coord = [tuple(map(int, line.split(","))) for line in lines]
num_boxes = len(coord)
dist = {}
for i in range(num_boxes):
    for j in range(i + 1, num_boxes):
        dist[(i,j)] = calc_dist(coord[i], coord[j])
sorted_dist = sorted(dist.items(), key=lambda item: item[1])
```

```python
circuits: list[set] = []
edges = [pair[0] for pair in sorted_dist[:10]]

while edges:
    circuit = set()
    connecting = True
    while connecting:
        connecting = False
        for edge in edges[:]:
            del_edge = False
            box1, box2 = edge
            if not circuit:
                circuit = {box1, box2}
                connecting = del_edge = True
            elif box1 in circuit:
                circuit.add(box2)
                connecting = del_edge = True
            elif box2 in circuit:
                circuit.add(box1)
                connecting = del_edge = True
            if del_edge:
                edges.remove(edge)
    circuits.append(circuit)
```

```python
clusters = sorted(circuits, key=len, reverse=True)
print(len(clusters[0]) * len(clusters[1]) * len(clusters[2]))
```

```python
full_circle = False
num_pairs = 0

while not full_circle:
    num_pairs += 1
    
    edges = [pair[0] for pair in sorted_dist[:num_pairs]]

    circuit = set()
    connecting = True
    while connecting:
        connecting = False
        for edge in edges[:]:
            del_edge = False
            box1, box2 = edge
            if not circuit:
                circuit = {box1, box2}
                connecting = del_edge = True
            elif box1 in circuit:
                circuit.add(box2)
                connecting = del_edge = True
            elif box2 in circuit:
                circuit.add(box1)
                connecting = del_edge = True
            if del_edge:
                edges.remove(edge)

    if len(circuit) == num_boxes:
        full_circle = True
```

```python
box1, box2 = sorted_dist[num_pairs-1][0]
score = coord[box1][0] * coord[box2][0]
print(score)
```
